```python
import pandas as pd

dim_mun = pd.read_excel('arquivos/RELATORIO_DTB_BRASIL_MUNICIPIO.xls', header=6)
```

```python
dim_mun
```

```python
# extrai os 6 primeiros dígitos do código do município
dim_mun['cd6_mun'] = dim_mun['Código Município Completo'].astype(str).str[:6]

# seleciona as colunas
dim_mun = dim_mun[['UF', 'Nome_UF', 'cd6_mun', 'Código Município Completo', 'Nome_Município']]

# renomeia as colunas
dim_mun.columns = ['cd_uf', 'nm_uf', 'cd6_mun', 'cd_mun', 'nm_mun']
```

```python
dim_mun
```

```python
dim_mun.to_csv('dimensoes/dim_mun.csv', index=False, encoding='latin1', sep=';')
```
