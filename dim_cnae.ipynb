```python
import pandas as pd

dim_cnae = pd.read_excel('arquivos/CNAE20_EstruturaDetalhada.xls', header=2)
```

```python
dim_cnae
```

```python
# seleciona as colunas
dim_cnae = dim_cnae[['Classe', 'Denominação']]

# exclui valores nulos
dim_cnae = dim_cnae.dropna(subset=['Classe'])

# padroniza a coluna de código
dim_cnae['Classe'] = dim_cnae['Classe'].astype(str).str.replace('.', '', regex=False).str[:4]

# renomeia as colunas
dim_cnae.columns = ['cd_classe', 'nm_classe']
```

```python
dim_cnae
```

```python
dim_cnae.to_csv('dimensoes/dim_cnae.csv', index=False, encoding='latin1', sep=';')
```
